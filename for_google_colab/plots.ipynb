```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
def plot(data):
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))
    
    ax1.plot(data['Epoch'], data['Loss'], marker='o', color='b')
    ax1.set_title('Loss vs Epochs')
    ax1.set_xlabel('Epochs')
    ax1.set_ylabel('Loss')
    
    ax2.plot(data['Epoch'], data['BAC'], marker='o', color='r')
    ax2.set_title('BAC vs Epochs')
    ax2.set_xlabel('Epochs')
    ax2.set_ylabel('BAC')
    
    plt.tight_layout()
    plt.show()
```

```python
def plot_loss_bac(data):
    fig, ax = plt.subplots(figsize=(10, 8))
    
    ax.plot(data['Epoch'], data['Loss'], marker='o', color='b', label='Loss')
    ax.plot(data['Epoch'], data['BAC'], marker='o', color='r', label='BAC')
    ax.set_title('Loss and BAC vs Epochs')
    ax.set_xlabel('Epochs')
    ax.set_ylabel('Value')
    
    ax.legend()
    
    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt

def plot_recall(data):
    plt.plot(data['Epoch'], data['Recall'], marker='o', color='b')
    plt.xlabel('Epoch')
    plt.ylabel('Recall')
    plt.title('Recall vs. Epoch')
    plt.xticks(data['Epoch']) 
    plt.show()
```
